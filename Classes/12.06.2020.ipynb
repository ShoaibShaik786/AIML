```python
# Coding Question Example

s= input()
ls1 =s.split(',')
ls2 = [i.split(':')[1] for i in ls1]
ls3 = []
for num in ls2:
    n = int(num)
    rev = num[::-1]
    while str(n) != rev:
        n += int(rev)
        rev = str(n)[::-1]
    ls3.append(n)
ls1[ls3.index(max(ls3))].split(':')[0]
        
```
