#for loops
 - for loop with range() function
 - for loop with iterable object

```
n = 10
for i in range(1,n+1):
  print(i,end=" ")

print()
i = 1
while i <= n:
  print(i,end=" ")
  i += 1
```

```
#for loop with iterable object

s = "pyhton"
for i in s:
  print(i)

#Using While loop

i = 0
while i < len(s):
  print(s[i],end=" ")  
  i += 1

#Using for loop but not with iterable object
for j in range(0,len(s)):
  print(s[j],end=" ")
```

1. Accept a string from the user and count no.of vowels,consonants and special characters

2. count the no.of pairs of a in the given string
    string --> "abbaaccbbaaa"
    output --> 2

3. string --> "((((()()()))))()"
   output --> 8



```
#Question 1
vcount = 0;  
ccount = 0;  
spcount = 0;
s= input("Enter a String: ")
   
s = s.lower() 
 
for i in range(0,len(s)):   
    if s[i] in ('a',"e","i","o","u"):  
        vcount = vcount + 1;  
    elif (s[i] >= 'a' and s[i] <= 'z'):  
        ccount = ccount + 1; 
    elif s[i] in ("+","-","/","*"):
      spcount = spcount + 1; 
  
print(vcount);  
print(ccount);
print(spcount)  
```
