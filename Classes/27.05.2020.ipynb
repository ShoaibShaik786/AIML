###Operators :Which perform some specific task
###Operands : Values given to operators 

```
#Arithmetic Operators: +,-,*,/,%,//(floor division),**(exponent)

a = int(input("Enter Value of a :"))
b = int(input("Enter Value of b :"))

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)#float quotient
print(a // b)#int quotient
print(a ** b)# power value
```

```
#Relational Operators: <,>,<=,>=,!=,==

a = int(input("Enter Value of a :"))
b = int(input("Enter Value of b :"))

print(a < b)
print(a > b)
print(a <= b)
print(a >= b)
print(a != b)
print(a == b)
```

```
#Logical operators: and,or,not

print(a < b and a > 1)
print(a > b or a > 1)
print(not a < b)
```

```
#Membership Operators : in,not in
list1 = [12,34,45,67,78,89]
print(12 in list1)
print(120 in list1)
print(12 not in list1)
print(120 not in list1)
```

```
#Conditional Statements : if, else, 

if a < b :
  print("a is less than b")
elif a > b :
  print("a is greater than b");
else :
  print("a is equal to b");
```
