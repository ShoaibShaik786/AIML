## Different types of arguments
  - Required positional arguments
  - Defaul arguments
  - Keyword argument
  - variable length argument

```
#Req. positional arguments
def add(a,b):
  return a+b
add(10,20)
```

```
def add(a,b=100):#default value(always should be present at end of list)
  return a+b
print(add(100))
print(add(100,200))#Actual value
```

```
#keyword arguments
def add(a,b):
  print("a = ",a)
  print("b = ",b)
add(a = "10",b = "20")
```

```
#variable length arguments
def add(*a):
  for arg in a:
    return sum(a)
print(add(100))
print(add(100,200))
print(add(100,200,300))
```

```
#variable length arguments
def mul(*a):
  m = 1;
  for i in a:
    m = m*i
  return m
print(mul(10,20))
print(mul(10,20,30))
print(mul(10,20,30,40))
```

##Nested Function

```
def square(a):
  return a*a
def cube(x):
  return square(x)*x
print(square(10))
print(cube(10))
```

##Recursion In Functions
 1. Factorial number
 2. Fibonacci series

```
#Factorial
def fact(n):
  if n == 1:
    return n
  else:
    return n*fact(n-1)
n = int(input())
if n > 0:
  print(fact(n))
```

```
#fibonacci
def fib(n):
  if n == 1:
    return 0
  elif n == 2:
    return 1
  else:
    return fib(n-1)+fib(n-2)

n = int(input("Enter a number : "))
print(fib(n))
```
