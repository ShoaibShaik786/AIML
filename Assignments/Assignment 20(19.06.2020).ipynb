```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('company_sales_data.csv')
df.head()
```

### Question 1

```python
df.describe()
```

### Question 2

```python
#Line Style dotted and Line-color should be red
#Show legend at the lower right location.
#X label name = month number
#Y label name = sold units number
#Add a circle marker.
#Line marker color as read
#Line width should be 3
```

```python
plt.title('monthly_profit')
plt.xlabel('month_number')
plt.ylabel('sold units number')
plt.plot(df['month_number'],df['total_units'],'-o',color = 'red',linestyle = 'dotted',linewidth = 3)
plt.legend({'total_units'},loc = 'lower right')
plt.savefig('plots\\monthly_profit.png')
plt.show()
```

### Question 3

```python
products = df.columns.to_list()[1:7] # all the product columns taken to list 

plt.figure(figsize=(10,6))
plt.xlabel("Months")
plt.ylabel("Units_Sold")
# using product list to plot 
for i in products:
    plt.plot(df["month_number"],df[i],"-o",label=i,lw=4)
plt.grid()
plt.title('Products Sold Per Month')
plt.legend(loc="best")
plt.savefig('plots\\Products_Sold.png')
plt.show()
```

### Question 4

```python
plt.title("Toothpaste Sales")
plt.xlabel("Month")
plt.ylabel("toothpaste sold(units)")
plt.scatter(df.month_number,df.toothpaste,color="red")
plt.savefig("plots\\toothpaste.png")
plt.show()
```

### Question 5

```python
plt.figure(figsize=(15,6))
plt.title("Face products")
plt.xlabel("Month number")
plt.ylabel("Units Sold")
plt.bar(df.month_number-0.2,df.facecream,label="facecream",width=0.4)
plt.bar(df.month_number+0.2,df.facewash,label="facewash",width=0.4)
plt.legend(loc="best")
plt.savefig("plots\\faceproducts")
plt.show()
```

### Question 6

```python
fig,a=plt.subplots(2) # Creating Objects for subplots and product data

a[0].plot(df.month_number,df[products[1]]) # Using Products list created earlier for selecting columns
a[0].set_title("Face Wash")
a[1].plot(df.month_number,df[products[3]])
a[1].set_title("Bathing Soap")

#padding between subplots 
fig.tight_layout(pad=3.0)
plt.savefig("plots\\soap.png")
plt.show()
```

### Question 7

```python
plt.figure(figsize=(15,6))
plt.hist(df.total_profit,bins=[50000,100000,150000,200000,250000,300000,350000,400000,450000,500000])
plt.title("total monthly profit")
plt.xlabel("Profit value")
plt.ylabel("no of months")
plt.grid()
plt.savefig("plots\profitRanges.png")
plt.show()
```
